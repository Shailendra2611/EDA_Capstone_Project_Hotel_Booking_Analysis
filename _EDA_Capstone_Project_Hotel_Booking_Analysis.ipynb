###Bookings by Months

````python
resort_guest = Resort_df['arrival_date_month'].value_counts().reset_index()
resort_guest.columns = ['month','no of guests']
resort_guest

city_guest = City_df['arrival_date_month'].value_counts().reset_index()
city_guest.columns = ['month','no of guests']
city_guest

#merge resort guest and city guest month wise
total_guest = resort_guest.merge(city_guest, on = 'month')
total_guest.columns = ['month','no of guests in resort','no of guest in city hotel']
total_guest
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
total_guest['month'] = pd.Categorical(total_guest['month'], categories=months, ordered=True)

#Which month get most visitors?
plt.figure(figsize =(15,8))
sns.lineplot(data = total_guest, x='month', y='no of guests in resort', marker ='^')
sns.lineplot(data = total_guest, x='month', y='no of guest in city hotel',marker ='^')
plt.legend(['Resort','City Hotel'])
plt.ylabel('Number of guest')
````

##### 1. Why did you pick the specific chart?

 The pair plot will show scatter plots for each pairwise combination of the selected variables, along with histograms for each individual variable. This plot is useful for visualizing relationships and distributions between variables in your dataset.

##### 2. What is/are the insight(s) found from the chart?

It is clear from the pair plot that the main diagonal of the pair plot represents the distribution of each variable individually. By examining these plots, you can quickly assess the shape of the distribution, whether it is symmetric (bell-shaped), skewed (asymmetric), or multimodal (multiple peaks). This information provides insights into the variable's central tendency and spread, allowing to understand its distributional characteristics.

The scatter plots in the upper and lower triangles of the pair plot show the relationship between each pair of variables. By analyzing these scatter plots, we can identify the general trend or pattern between two variables. For example, if the points in the scatter plot form a clear upward or downward trend, it indicates a positive or negative relationship, respectively.

Moreover, we can assess the strength of the relationship between variables by examining the dispersion of the points in the scatter plots. If the points are closely clustered around a line, it suggests a strong correlation, while a more scattered distribution indicates a weaker correlation or no significant relationship.
